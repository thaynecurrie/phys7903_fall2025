```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from matplotlib.ticker import AutoMinorLocator
hip99770b=np.loadtxt('hip99770b_spec_with_errors.txt')
hd33632ab=np.loadtxt('hd33632ab_spec_with_errors.txt')

hip99770bwvlh=hip99770b[:,0]
hip99770bflux=hip99770b[:,1]
hip99770beflux=hip99770b[:,2]

hd33632abwvlh=hd33632ab[:,0]
hd33632abflux=hd33632ab[:,1]
hd33632abeflux=hd33632ab[:,2]

#scale HD 33632 Ab's flux density to HIP 99770 b

scale=np.mean(hip99770bflux/hd33632abflux)

hd33632abflux*=scale

fig,ax=plt.subplots(1,2,figsize=(24,8))

for i,axes in enumerate(ax.flat):

 if i == 0:
  axes.plot(hip99770bwvlh,hip99770bflux, marker='o',lw=3,color='tab:green', label='HIP 99770 b')
  axes.fill_between(hip99770bwvlh,hip99770bflux-hip99770beflux,hip99770bflux+hip99770beflux,
                   alpha=0.4,color='tab:green')

  axes.plot(hd33632abwvlh,hd33632abflux, marker='o',lw=3,color='magenta', label='HD 33632 Ab (scaled)')
  axes.fill_between(hd33632abwvlh,hd33632abflux-hd33632abeflux,hd33632abflux+hd33632abeflux,
                   alpha=0.2,color='magenta')


#axes.plot(hip99770bwvlh,hip99770bflux, marker='o')
#axes.fill_between(hip99770bwvlh,hip99770bflux-hip99770beflux,hip99770bflux+hip99770beflux,alpha=0.3)
#plt.

#plt.plot(hip99770bwvlh,hip99770bflux,lw=5,marker='-o')
 scale=1
    
 if i > 0:
    
  axes.errorbar(hip99770bwvlh,hip99770bflux, hip99770beflux,marker='o',lw=3,color='tab:green', label='HIP 99770 b')
         #    alpha=0.4,color='tab:green')
  #axes.fill_between(hip99770bwvlh,hip99770bflux-hip99770beflux,hip99770bflux+hip99770beflux,
                   

  axes.errorbar(hd33632abwvlh,hd33632abflux, hd33632abeflux,marker='o',lw=3,color='magenta', label='HD 33632 Ab (scaled)')
  #axes.fill_between(hd33632abwvlh,hd33632abflux-hd33632abeflux,hd33632abflux+hd33632abeflux,
           #        alpha=0.2,color='magenta') 
    
 axes.set_ylabel(r'Flux Density (mJy)',fontsize=20)
 axes.set_xlabel(r'Wavelength ($\mu m$)',fontsize=20)
 axes.tick_params(which='both',direction='in',labelsize=14*scale)
 axes.tick_params(which='major',length=10,width=1.5*scale)
 axes.tick_params(which='minor',length=5,width=1.5*scale)
 axes.xaxis.set_minor_locator(AutoMinorLocator(5))
 axes.yaxis.set_minor_locator(AutoMinorLocator(5))
 axes.legend(loc='upper left',fontsize=14)

#thicken the spines
 for axl in ['top','bottom','left','right']:
  axes.spines[axl].set_linewidth(2)
```
